````python
import collections

def analyzeThis(text):
    text = text.lower()
    ftext = u''.join([t for t in text.lower() if t.isalnum() or t == ' '])

    words = collections.Counter(ftext.split())
    symbols = collections.Counter(ftext.replace(' ',''))
    
    sent = text.count('.') + text.count('!') + text.count('?')
    print u"Предложений %d, слов %d, уникальных слов %d, букв %d" % (sent, 
                                                                     sum(words.values()), 
                                                                     len(words.keys()), 
                                                                     sum(symbols.values()),
                                                                     )
    
    return symbols
    
    
def capitalist(text):
    ret = []
    state = 1
    for x in text:
        t = x
        if state == 1 and x.isalnum():
            t = x.upper()
            state = 0
        if state == 0 and x in '.?!':
            state = 1
        ret.append(t)
    return ''.join(ret)
        
    
text = u"""
Тут будут разные знаки: препинания и 
обступы. 

Возможно, даже с ошибками. некоторые с маленькой буквы. 
какие-то со знаком вопроса! или восклицания?
"""

textstat = analyzeThis(text)

print capitalist(text)
````

### Задача о матане

Насколько я понимаю условие, требуется найти такое $k^{*}(\lambda, \mu)$, что для всех $k > k^{*}$ будет верно $P(k) = \frac{\lambda^k}{k!}e^{-\lambda} \le \mu$.

````python
def poissonP(param, num=100):
    k = np.arange(1, num)
    t = param / k.astype('float')
    #just a dirty trick avoiding zero division
    t = np.insert(t, 0, 1)
    k = np.insert(k, 0, 0)
    return (k, np.cumprod(t) * np.exp(- param))

c = iter(['green', 'blue', 'red'])

for p in [80.0, 30.0, 0.5]:
    plt.scatter(*poissonP(p), marker=".", color=c.next())

plt.xlim([-1, 101])
plt.show()
````

Для определенности, пусть речь идет про $\lambda \gg 1$. 
Оценим вероятность выпадения наиболее вероятного числа:
$$\mu^{*} = P(k = \lambda) = \frac{\lambda^k}{k!}e^{-\lambda} \approx \frac{1}{\sqrt{2\pi\lambda}}$$

*мы воспользовались формулой Стирлинга $n! \approx (n/e)^n \sqrt{2\pi n}$*


Поскольку распределение похоже на нормальное со средним $\mu = \lambda$ и дисперсией $\sigma^2 = \lambda$, то можно попробовать записать ответ сравнительно аккуратно через функцию ошибок. К сожалению, теорвер слишком забыт, чтобы это делать уверено.
Поэтому напишем пару простых функций.
Для практического применения необходимо несколько сузить область интереса.

Если решение будет использоваться для $\mu \sim \mu^{*}$, скажем для $\frac{\mu}{\mu^{*}} \ge 0.1$, можно воспользоваться обратной функцией `revPoisson(`$\lambda$`)`.
Она построена довольно просто -- строим табличку для фиксированного $\lambda$, по однозначному участку интерполируем.

Безусловно решение можно причесать: выбирать количество точек автоматом по параметру распределения, возвращать целочисленные индексы и т.д.

````python
def revPoisson(p, num=100):
    k = np.arange(1, num)
    t = p / k.astype('float')
    #just a dirty trick avoiding zero division
    t = np.insert(t, 0, 1)
    k = np.insert(k, 0, 0)
    t = np.cumprod(t) * np.exp(- p)
    idx = np.argmax(t)
    print idx, t[idx]
    func = interp1d(t[idx:], k[idx:])
    
    #just illustration of approach:
    plt.scatter(t, k, marker='.')
    plt.plot(t, func(t), 'r')
    plt.xlabel(r"$P$")
    plt.ylabel(r"$k$")
    plt.show()
    #
    
    return func


rev = revPoisson(10.0)
print rev([0.1, 0.05, 0.01])
````
